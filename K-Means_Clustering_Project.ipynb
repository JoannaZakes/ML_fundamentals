# K Means Clustering Project 

Using KMeans Clustering to cluster Universities into to two groups: Private and Public.

___
It is **very important to note, there are labels in this data set, but they will NOT be used for the KMeans clustering algorithm, since that is an unsupervised learning algorithm.** 

Usually the Kmeans algorithm is used in the case the labels aren't available. In this case the labels will be used to try to get an idea of how well the algorithm performed, but it won't usually be possible for Kmeans and it doesn't make sense in a real world setting!.

## K Means Cluster Creation


```python
from sklearn.cluster import KMeans
```

**Creating an instance of a K Means model with 2 clusters**

```python
kmeans = KMeans(n_clusters=2)
```

**Fitting the model to all the data except for the Private label.**

```python
kmeans.fit(colleges.drop('Private', axis=1))
```

**Cluster center vectors:**

```python
kmeans.cluster_centers_
```

## Evaluation

There is no perfect way to evaluate clustering if you don't have the labels, however since this is just an exercise, we do have the labels, so we take advantage of this to evaluate our clusters.

**Creating a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.**

```python
colleges['Cluster'] = colleges['Private'].apply(lambda x: 1 if x=='Yes' else 0)
```

```python
colleges.head(1)
```

**Creating a new column for df called 'KMeans_cluster', which represents the 2 groups identified by the algorithm**

```python
colleges['KMeans_cluster'] = kmeans.labels_
```

```python
colleges.head(1)
```

**Scatterplot of F.Undergrad versus Outstate where the points are colored by original label (original) and predicted label (kmeans)**

```python
fig, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title('original')
ax1.scatter(data=colleges, x='Outstate', y='F.Undergrad', c='Cluster', cmap='rainbow')

ax2.set_title('kmeans')
ax2.scatter(data=colleges, x='Outstate', y='F.Undergrad', c='KMeans_cluster', cmap='rainbow')
```
